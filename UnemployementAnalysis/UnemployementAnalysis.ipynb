## PROJECT-2 : UNEMPLOYMENT ANALYSIS USING PYTHON DATA SCIENCE 
Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, analyzing the unemployment rate can be done using data science.

The following project comprises the usage of various plots and graphs to analyse the data gievn in the dataset.

## VISUALISATION OF DATA : CSV FILE -> "UNEMPLOYMENT IN INDIA"

```python
plt.figure(figsize=(12, 10))
sn.histplot(d[" Estimated Unemployment Rate (%)"], kde=True)
plt.xlabel("Unemployment Rate (%)")
plt.title("Distribution of Unemployment Rate")
plt.grid(True)
plt.show()
```

The above histogram depicts which Unemployement Rate category had the highest count among different regions.
It can be observed that most states unemployement rate lies between 0-10%

```python
avg_rate = d.groupby(["Region", "Date"])[" Estimated Unemployment Rate (%)"].mean().unstack()
colors = plt.cm.get_cmap('tab20', avg_rate.T.shape[1])
fig, ax = plt.subplots(figsize=(20, 15))
avg_rate.T.plot(kind='area', stacked=True, alpha=0.6, ax=ax, colormap=colors)

plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")
plt.title("Average Unemployment Rate Trend Across States (Area Plot)")
plt.legend(title="State", loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

The above area plot depicts unemployment rate of each state with time. It can be infered that highest unemployement rate was around March-May and in states like Uttrakhand, West Bengal, and Uttar Pradesh.

```python

for state in d["Region"].unique():
  # Filtering data for current state (consider potential missing "Area" column)
  state_data = d[d["Region"] == state]
  rural_data = state_data[state_data["Area"] == "Rural"]
  urban_data = state_data[state_data["Area"] == "Urban"]
  
  
  
  # Plotting lines with labels for rural and urban (adjust labels as needed)
  plt.figure(figsize=(10, 6))  
  plt.plot(rural_data["Date"], rural_data[" Estimated Unemployment Rate (%)"], marker='o', markersize=8, label="Rural", color='midnightblue')
  plt.plot(urban_data["Date"], urban_data[" Estimated Unemployment Rate (%)"], marker='s', markersize=8, label="Urban", color='cyan',markeredgewidth=1, markeredgecolor='black')
  
  
      

  # Customizing plot appearance 
  plt.xlabel("Date")
  plt.ylabel("Unemployment Rate (%)")
  plt.title(f"Unemployment Rate Trend in {state}")  
  plt.grid(True)
  plt.legend(title="Area") 
  plt.xticks(rotation=35)  
  plt.tight_layout()  
  plt.show()  
```

The above code is used to get line plots with trends of unemployment with time in rural and urban areas of each state respectively. Insightful information can hence be drawn from these plots.

```python
# Calculate total unemployment rate for rural and urban areas
rural_unemployment = d[d["Area"] == "Rural"][" Estimated Unemployment Rate (%)"].mean()
urban_unemployment = d[d["Area"] == "Urban"][" Estimated Unemployment Rate (%)"].mean()

# Create a bar chart
plt.figure(figsize=(6, 6))
plt.bar(["Rural", "Urban"], [rural_unemployment, urban_unemployment])
plt.xlabel("Area")
plt.ylabel("Average Unemployment Rate (%)")
plt.title("Unemployment Rate by Area")
plt.grid(axis='y')
plt.show()
```

Above bar plot depicts average unemployment rates in all rural and urban region of all states combined. It can be infered that urban areas had a higher average rate of unemployment as compared to rural areas.

```python
plt.figure(figsize=(10, 10))
plt.scatter(d[" Estimated Labour Participation Rate (%)"], d[" Estimated Unemployment Rate (%)"], marker='o',edgecolors='black',facecolors='cyan',linewidth=1)
plt.xlabel("Estimated Labour Participation Rate (%)")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.title("Labour Participation Rate vs Unemployment Rate")
plt.grid(True)
plt.show()
```

From above scatter plot between Estimated Labour Participation Rate and Estimated Unemployment rate; we can infer that Labour Participation for majority of regions is around 30-50% and unemployement rate is between 0-25%

```python
# Pivot table to get unemployment rate for each state and date
unemployment_pivot = d.pivot_table(index="Region", columns="Date", values=" Estimated Unemployment Rate (%)")

# Creating a heatmap
plt.figure(figsize=(20, 20))
sn.heatmap(unemployment_pivot, cmap="YlGnBu")  
plt.xlabel("Date")
plt.ylabel("State")
plt.title("Unemployment Rate Across States")
plt.show()
```

The above heatmap is between dates and states.
The color intensity of each cell indicates the magnitude of the unemployment rate, with darker shades representing higher unemployment rates and lighter shades representing lower unemployment rates.
It can thus be infered based on the above plot that between 2020-05-30 and 2020-06-30 Puducherry saw the highest unemployment rate.

```python
correlation_matrix = df[[ " Estimated Labour Participation Rate (%)", " Estimated Employed", " Estimated Unemployment Rate (%)"]].corr()
print(correlation_matrix)
```

```python
# Grouping data by State, Date and get average unemployment rate for each date
avg_rate = df.groupby(["Region", "Date"])[" Estimated Unemployment Rate (%)"].mean().unstack()

# Plotting line chart with multiple lines for each state
avg_rate.plot(kind='line', figsize=(20, 9))
plt.xlabel("State")
plt.ylabel("Average Unemployment Rate (%)")
plt.title("Average Unemployment Rate Trend Across States")
plt.legend(title="State")
plt.grid(True)
plt.show()
```

The above line plot is between states and Average unemployment rates on different dates; different dates being represented by lines of different colors.
It can be infered that on 2020-04-30, unemployment was highest , around 75 % near regions like Meghalaya, Sikkim; and lowest around 4-5% near regions of Goa and Jharkhand.
Similarly, inferences can be made for other dates as well.

```python
# Plot between states and mean employed values
avg_rate = df.groupby(["Region"])[" Estimated Employed"].mean()

# Plotting line chart with multiple lines for each state
custom_colors = [
                "blue", "orange", "green", "red", "purple", "brown", "pink", "gray", "yellow",
                "cyan", "lightblue", "magenta", "lightgreen", "maroon", "lavender", "indigo", "lightpink", "lightgray",
                "lightyellow", "lightcyan", "darkblue", "darkgreen", "darkorange", "darkred", "midnightblue", "chocolate", "hotpink"
              ]


avg_rate.plot(kind="bar", figsize=(20, 10), color=custom_colors)
plt.xlabel("State")
plt.ylabel("Estimated Employed")
plt.title("Estimated Employed Trend Across States")
plt.grid(True)
plt.show()
```

The above bar graph is plotted between states and estimated employed values in those states.
It can be infered based on the above figure that amongst all states, Uttar Pradesh had the highest number of employement values and regions of Goa, Sikkim and Puducherry had the lowest employed.
